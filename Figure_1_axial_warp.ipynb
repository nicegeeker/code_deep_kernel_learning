# Axial Transformation Unit

```python
res = 11
grad = 30
llim = 0
rlim = 1
theta_grad = torch.ones((res - 1, 1)) * torch.tensor(grad)
# theta_locs = torch.linspace(llim, rlim, steps=(res - 1) + 2)[1:-1].reshape(res - 1, 1)
theta_locs = torch.linspace(llim, rlim, steps=10)
```

```python
x = torch.linspace(0, 1, 1000)
```

```python
y = np.zeros((10, 1000))
for i in range(10):
    y[i, :] = 1 / (1 + np.exp(-theta_grad[i, 0] * (x - theta_locs[i])))
```

```python
ATU_1 = AxialWarpUnit(dim=0, res=10, grad=30, lims=[0, 1.0])
ATU_2 = AxialWarpUnit(dim=1, res=10, grad=30, lims=[-1, 0])
```

```python
with plt.style.context(["science", "ieee", "no-latex", "grid"]):
    fig = plt.figure(constrained_layout=True, figsize=(3.3, 3.3))
    gs = GridSpec(3, 2, figure=fig)
    ax1 = fig.add_subplot(gs[0:2, :])
    ax2 = fig.add_subplot(gs[2, 0])
    ax3 = fig.add_subplot(gs[2, 1])
    #     fig, ax = plt.subplots(figsize=(7.16, 2.5))
    ax1.plot(x, x)
    for i in range(10):
        ax1.plot(x, y[i, :])
    ax1.set_title("(a) Basis functions")
    ax1.set_xlabel(r"$s_k$")
    ax1.set_ylabel(r"$\phi(s_k)$")
    oned_map(ATU_1, ax=ax2)
    oned_map(ATU_2, ax=ax3, dim=1)
    ax2.set_title("(b) Dimension 1")
    ax2.set_xlabel(r"$s_1$")
    ax2.set_ylabel(r"$f(s_1)$")
    ax3.set_title("(c) Dimension 2")
    ax3.set_xlabel(r"$s_2$")
    ax3.set_ylabel(r"$f(s_2)$")
save_fig("axial_transformation_unit")
```
