```python
import numpy as np

from matplotlib import pyplot as plt


%matplotlib inline
```

```python
def get_sigmoid_para(grad, res, lims):
        theta = np.zeros((2, res-1))
        theta[0, :] = grad
        theta[1, :] = np.linspace(lims[0], lims[1], num=(res-1) + 2)[1:-1]
        return theta
```

```python
theta = get_sigmoid_para(100, 11, [0, 1])
```

```python
theta
```

```python
x = np.linspace(0, 1, 1000)
```

```python
y = np.zeros((10, 1000))
for i in range(10):
    y[i, :] = 1/(1+ np.exp(-theta[0,i] * (x-theta[1,i])))

```

```python
y
```

```python
fig, ax= plt.subplots(nrows=1, ncols=1, figsize = (16,4))
for i in range(10):
    ax.plot(x, y[i,:])
```
